```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_excel('/content/b21_kor_v2.5.xlsx')
df = df.dropna()
```

```python
y_data = df['재산피해금액']
x_data =df.drop(['재산피해금액'],axis=1,inplace=False)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=156)
x_test, x_val, y_test, y_val = train_test_split(x_test, y_test, test_size=0.5, random_state=156)

print('train shape:', x_train.shape, y_train.shape)
print('test shape:', x_test.shape, y_test.shape)
print('val shape:', x_val.shape, y_val.shape)
```

```python
model = Sequential()
model.add(Dense(64, activation = 'relu', input_shape = (132, )))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss = 'mse', metrics=['mae'])

history = model.fit(x_train, y_train, epochs=1500, validation_data=(x_val, y_val))
model.evaluate(x_test, y_test) #오차가 더 줄어들지 않으면 과적합 되었다고 판단
```

```python
results = model.predict(x_test).reshape(110,)
```

```python
for i in range(1,20):
  print(f'pred: {results[i]}')
  print(f'y_test:{y_test.iloc[i]}\n')
```
