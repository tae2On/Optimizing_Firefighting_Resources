```python
b21 = pd.read_excel('/content/b21_kor.xlsx')
```

```python
warnings.filterwarnings('ignore')

# 건물 정보 기준
df_buld = b21[[
'건물구조식',
'건물구조조',
'건물구조즙',
'건물구조동수',
'지상층수',
'지하층수',
'연면적',
'바닥면적',
'건물상태명'
]]
df_buld_corr = df_buld.corr()


# 피해 기준
df_dmge = b21[[
'사망수',
'부상자수',
'사망인명피해수',
'재산피해금액'
]]
df_dmge_corr = df_dmge.corr()


# 수행 시간
df_time = b21[[
'출동소요시간',
'화재진압시간'
]]
df_time_corr = df_time.corr()


# 지역별
df_loc1 = b21[[
#'읍면동명',
'도시농촌구분명',
'읍면동구분명'
]]
df_loc1_corr = df_loc1.corr()


# 좌표별
df_loc2 = b21[[
'GIS_X좌표',
'GIS_Y좌표',
'경도',
'위도'
]]
df_loc2_corr = df_loc2.corr()


# 거리별
df_dist = b21[[
'현장소방서거리',
'현장안전센터거리',
'현장소방지역대거리'
]]
df_dist_corr = df_dist.corr()


# 화재요인별
df_resn = b21[[
'발화열원명',
'발화요인대분류명',
'최초착화물대분류명',
'발화기기대분류명',
'연소확대물대분류명',
'시설장소대분류명',
'발화층명',
'그을음면적'
]]
df_resn_corr = df_resn.corr()


# 법규 관련
df_law = b21[[
'화재보험의무가입대상여부',
'방화관리대상여부',
'다중이용업여부',
'긴급통제단운영여부'
]]
df_law_corr = df_law.corr()


# 환경 정보
df_env = b21[[
'시간단위기온',
'시간단위강수량',
'시간단위풍속',
'시간단위풍향',
'시간단위습도',
'시간단위적설량'
]]
df_env_corr = df_env.corr()
```

```python
def corr_filter(df):
  corr_lst = []
  for col in df.columns:
    for idx in df.index:
      if df[col][idx] >= 0.3 and df[col][idx] < 1:
        print(col, idx, df[col][idx])
```

```python
# 건물 정보 기준
corr_filter(df_buld_corr)
```

```python
# 피해 기준
corr_filter(df_dmge_corr)
```

```python
# 수행 시간
corr_filter(df_time_corr)
df_time_corr
```

```python
# 지역별
corr_filter(df_loc1_corr)
df_loc1_corr
```

```python
# 좌표별
corr_filter(df_loc2_corr)
```

```python
# 거리별
corr_filter(df_dist_corr)
```

```python
# 화재요인별
corr_filter(df_resn_corr)
df_resn_corr
```

```python
# 법규 관련
corr_filter(df_law_corr)
df_law_corr
```
